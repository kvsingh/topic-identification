```python
#code for checking how many top 10 tokens of songs does this token come in.
count_occurs_in = {}
for doc in corpus:
    this_bow = sorted(doc, key=lambda a:a[1], reverse=True)
    for (tokenid, count) in this_bow[:10]:
        count_occurs_in[tokenid] = count_occurs_in.get(tokenid, 0) + 1

count_occurs_in_items = sorted(count_occurs_in.items(), key=lambda a:a[1], reverse=True)
for (tokenid, count) in count_occurs_in_items[:10]:
    print inverted_dict[tokenid], count
```
